```python
x=data.drop(['Class'],axis=1)
y=data['Class']
x.shape,y.shape
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
x_1, y_1 = smote.fit_resample(x, y)

print(x_1.shape, y_1.shape)
```

```python
plt.figure(figsize=(10,6))

sns.countplot(x=y_1,palette=['blue','red']);
```

##Feature Selection & Engineering
Identify relevant features for the model and drop irrelevant features.
**Normalize the numerical features.**

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_1['Amount'] = scaler.fit_transform(x_1['Amount'].values.reshape(-1, 1))
```

```python
corr=x_1.corr()
plt.figure(figsize=(24,20))
sns.heatmap(corr, cmap='coolwarm', annot=True)
plt.show()
```

**Engineer new features that may be useful for the model, such as time-based features.**

```python
delta_time = pd.to_timedelta(x_1['Time'], unit='s')
x_1['Time_Day'] = (delta_time.dt.components.days).astype(float)
x_1['Time_Hour'] = (delta_time.dt.components.hours).astype(float)
x_1['Time_Minute'] = (delta_time.dt.components.minutes).astype(float)
x_1.tail(100)
```

**By Experimenting with different Time features and their impact on model's performance below features are dropped**

```python
x_1.drop(['Time'],axis=1,inplace=True)
x_1.head()
```

##Model Selection

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_1,y_1,test_size=0.3,random_state=40)
```

LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
model1=LogisticRegression()
model1.fit(x_train,y_train)
```

accuracy score on train data

```python
from sklearn.metrics import accuracy_score
x_train_pred = model1.predict(x_train)
accuracy_score(y_train, x_train_pred)
```

accuracy score on test data

```python
x_test_pred = model1.predict(x_test)
accuracy_score(y_test, x_test_pred)
```

```python
model1.score(x_test,y_test)
```

```python
y_pred=model1.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

##**Evaluation Metrics**

**Calculate performance metrics such as accuracy, precision, recall, F1-score, and AUC-ROC.**

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
auc_roc = roc_auc_score(y_test, y_pred)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
print("AUC-ROC:", auc_roc)
```

##**Confusion Matrix**
**Plot and analyze the confusion matrix to understand the model's performance on different classes.**

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Compute Confusion Matrix
conf_mat = confusion_matrix(y_test, y_pred)

# Visualize Confusion Matrix
disp = ConfusionMatrixDisplay(conf_mat)
disp.plot()
```

```python
from sklearn.metrics import roc_curve, auc

# Calculate ROC curve and AUC
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```
